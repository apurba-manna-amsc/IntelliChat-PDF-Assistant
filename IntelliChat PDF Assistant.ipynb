```python
from google.colab import userdata
import os

# Fetch Hugging Face API Token
sec_key = userdata.get("HUGGINGFACEHUB_API_TOKEN")

# Set the environment variable for Hugging Face API token
os.environ["HUGGINGFACEHUB_API_TOKEN"] = sec_key
```

```python
from langchain_huggingface import HuggingFaceEndpoint

# Hugging Face model setup
repo_id = "mistralai/Mistral-7B-Instruct-v0.2"
llm = HuggingFaceEndpoint(repo_id=repo_id, max_length=128, temperature=0.7, token=sec_key)
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# Initialize embedding model (replace with the desired model)
embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")

# Generate embeddings for a test string
embeddings = embed_model.get_text_embedding("Hello World!")
print(f"Embedding size: {len(embeddings)}")
print(f"Embedding preview: {embeddings[:5]}")
```

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader

# Load documents from the directory (ensure the 'Data' folder exists and contains files)
documents = SimpleDirectoryReader("Data").load_data()

# Create the index from the documents
index = VectorStoreIndex.from_documents(documents)
```

```python
# Create query engine
query_engine = index.as_query_engine()
```

```python
# Sample queries
response_1 = query_engine.query("What is the capital of India?")
print(response_1.response)

response_2 = query_engine.query("What are student support services guidelines?")
print(response_2.response)
```

```python
response_2 = query_engine.query("what is ai chef master")
print(response_2)
```

```python
import gradio as gr
def get_answer(query):
    # Query the chatbot and get the answer
    response = query_engine.query(query)
    response = response.response
    if "does not provide" in response.lower():
        response = "For more information, please contact helpdesk@aichefmaster.com"

    return response
iface = gr.Interface(fn=get_answer, inputs="text", outputs="text", live=True,
                     title="PDF Chatbot", description="Ask questions based on the content of the PDF.")
iface.launch(share=True)
```
